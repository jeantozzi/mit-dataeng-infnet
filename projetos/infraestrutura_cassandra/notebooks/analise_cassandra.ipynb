```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pyspark.sql.functions import when, col, sum as spark_sum
import os
from pyspark.sql import SparkSession
import findspark
findspark.init()
```

```python
# Inicializando a sessão do Spark com integração ao Cassandra
spark = SparkSession.builder \
    .appName("AnaliseAcidentes") \
    .config("spark.jars.packages", "com.datastax.spark:spark-cassandra-connector_2.12:3.5.1") \
    .config("spark.cassandra.connection.host", "cassandra-1") \
    .config("spark.cassandra.connection.port", "9042") \
    .getOrCreate()

# Carregando a tabela no Spark DataFrame
df = spark.read \
    .format("org.apache.spark.sql.cassandra") \
    .options(table="acidentes", keyspace="datatran") \
    .load()

df.show()
```

# Transformando e exportando os dados

```python
# Adicionando a coluna 'tipo_dia' para identificar fim de semana e dia útil
df = df.withColumn(
    'tipo_dia',
    when(col('dia_semana').isin('sábado', 'domingo'), 'Fim de Semana').otherwise('Dia Útil')
)

# Filtrando para manter apenas os casos de "Com Vítimas Fatais"
df_fatais = df.filter(col('classificacao_acidente') == "Com Vítimas Fatais")

# Agrupando por ano, tipo_dia, fase_dia, condicao_metereologica e somar as mortes
df_agregado = df_fatais.groupBy('ano', 'tipo_dia', 'fase_dia', 'condicao_metereologica') \
    .agg(spark_sum('mortos').alias('mortos')).orderBy('ano', 'tipo_dia', 'fase_dia', 'condicao_metereologica')

df_agregado.show()

# Exportando o arquivo, combinando partições em um único arquivo CSV
os.makedirs("/output", exist_ok=True)
df_agregado.coalesce(1).write.csv('/output/acidentes_com_mortos_2017_2024', header=True, mode='overwrite')
```

# Criando os gráficos comparativos

```python
# Transformando o Spark Dataframe em Pandas Dataframe, para facilitar a criação das análises
df_pandas = df_agregado.toPandas()
```

```python
# Agrupando pela média de 'mortos' para cada combinação de 'ano' e 'tipo_dia'
df_media_mortos_por_tipo_dia = df_pandas.groupby(['ano', 'tipo_dia'], as_index=False)['mortos'].mean().round(2)

df_media_mortos_por_tipo_dia.head()
```

```python
# Criando o gráfico
plt.figure(figsize=(10, 6))

# Usando uma paleta de cores distinta para os anos
sns.barplot(data=df_media_mortos_por_tipo_dia, x='tipo_dia', y='mortos', hue='ano', palette='Set1')

# Adicionando rótulos de dados
for p in plt.gca().patches:
    height = p.get_height()
    if height > 0:
        plt.gca().annotate(f'{height:.2f}', 
                           (p.get_x() + p.get_width() / 2., height), 
                           ha='center', va='center', 
                           fontsize=10, color='black', 
                           xytext=(0, 5), textcoords='offset points')

# Ajustes finais
plt.title('Média de Mortos por Tipo de Dia e Ano')
plt.xlabel('Tipo de Dia')
plt.ylabel('Média de Mortos')
plt.legend(title='Ano', loc='upper right')

# Exibindo gráfico
plt.tight_layout()
plt.show()
```

```python
# Agrupando pela média de 'mortos' para cada combinação de 'ano' e 'fase_dia'
df_media_mortos_por_fase_dia = df_pandas.groupby(['ano', 'fase_dia'], as_index=False)['mortos'].mean().round(2)

df_media_mortos_por_fase_dia.head()
```

```python
# Definindo a ordem das fases do dia
ordem_fases_dia = ['Amanhecer', 'Pleno dia', 'Anoitecer', 'Plena Noite']

# Convertendo a coluna 'fase_dia' para uma categoria ordenada
df_media_mortos_por_fase_dia['fase_dia'] = pd.Categorical(df_media_mortos_por_fase_dia['fase_dia'], categories=ordem_fases_dia, ordered=True)

# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))

# Plotando o gráfico de dispersão com base em 'fase_dia' e 'ano', com 'mortos' como o eixo y
sns.scatterplot(data=df_media_mortos_por_fase_dia, x='fase_dia', y='mortos', hue='ano', palette='Set1', s=100, marker='o')

# Adicionando título e rótulos
plt.title('Média de Mortos por Fase do Dia e Ano', fontsize=14)
plt.xlabel('Fase do Dia', fontsize=12)
plt.ylabel('Média de Mortos', fontsize=12)

# Ajustando a legenda para o canto inferior direito
plt.legend(title='Ano', loc='lower right')

# Adicionando o guia do eixo Y (linhas pontilhadas)
plt.grid(True, axis='y', linestyle='--', color='gray', alpha=0.7)  # Linhas pontilhadas no eixo y

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

```python
# Agrupando pela média de 'mortos' para cada combinação de 'ano' e 'condicao_metereologica'
df_media_mortos_por_condicao_metereologica = df_pandas.groupby(['ano', 'condicao_metereologica'], as_index=False)['mortos'].mean().round(2)

df_media_mortos_por_condicao_metereologica.head()
```

```python
# Agrupando pela média de 'mortos' para cada combinação de 'ano' e 'condicao_metereologica'
df_media_mortos_por_condicao_metereologica = df_pandas.groupby(['ano', 'condicao_metereologica'], as_index=False)['mortos'].mean().round(2)

# Aplicando logaritmo (log(1 + x)) nas médias de mortos
df_media_mortos_por_condicao_metereologica['mortos'] = np.log1p(df_media_mortos_por_condicao_metereologica['mortos'])

# Calculando a média geral por condição meteorológica (todos os anos)
df_media_mortos_geral = df_media_mortos_por_condicao_metereologica.groupby('condicao_metereologica', as_index=False)['mortos'].mean()

# Criando o subgráfico para o gráfico de média geral
fig, ax = plt.subplots(figsize=(10, 6))

# Plotando o gráfico de barras comparando a média geral com a ordem definida
sns.barplot(data=df_media_mortos_geral, x='condicao_metereologica', y='mortos', ax=ax)

# Adicionando título e rótulos para o gráfico de comparação geral
ax.set_title('Média de Mortos por Condição Meteorológica (Todos os Anos)', fontsize=12)
ax.set_xlabel('Condição Meteorológica', fontsize=10)
ax.set_ylabel('Log(Média de Mortos)', fontsize=10)
ax.tick_params(axis='x', rotation=45)

# Ajustando para o layout
plt.tight_layout()
plt.show()
```
